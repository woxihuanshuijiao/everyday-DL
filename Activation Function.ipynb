```python
import torch
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
a = [[1, 0, -1, 5],
     [5, -1, 0, 1]]

a_np = np.asarray(a)

plt.imshow(a_np)
plt.show()
```

```python
relu = torch.nn.ReLU()

a_np = torch.from_numpy(a_np).float().unsqueeze(dim=0)
a_relu = relu(a_np)

print(a_relu)
plt.imshow(a_relu.detach().numpy().squeeze())
```

```python
b = [[1, 0, -1],
     [-1, 0, 1]]

b_np = np.asarray(b)

plt.imshow(b_np, cmap='gray')
plt.show()
```
