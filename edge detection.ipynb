```python
# 导入需要的库   
# 没有下载openCv请在终端输入：pip install opencv-python -i https://pypi.douban.com/simple
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
# 画图函数
def img_show(title, image):
    plt.title(title)
    plt.imshow(image, cmap="gray")
    plt.show()
```

```python
# 读入图片 
# 边缘检测算法需要灰度图
img = cv.imread("1.jpg", cv.IMREAD_GRAYSCALE)
img_show("grey-cabbage", img)
```

```python
# Sobel算子
# cv2.CV_64F防止梯度变为负，信息丢失

k_size = 3

# x, y方向
sobelx = cv.Sobel(img, cv.CV_64F, 1, 0, ksize = k_size)
sobely = cv.Sobel(img, cv.CV_64F, 0, 1, ksize = k_size)

sobel_X = cv.convertScaleAbs(sobelx)
sobel_Y = cv.convertScaleAbs(sobely)

# 两张图片以一定权重混合
img_sobel = cv.addWeighted(sobel_X, 0.5, sobel_Y, 0.5, 0)

img_show("sobel-cabbage", img_sobel)
```

```python
# scharr算子

scharr_x = cv.Scharr(img, cv.CV_8U, 1, 0)
scharr_y = cv.Scharr(img, cv.CV_8U, 0, 1)
scharrX = cv.convertScaleAbs(scharr_x)
scharrY = cv.convertScaleAbs(scharr_y)

img_scharr = cv.addWeighted(scharrX, 0.5, scharrY, 0.5, 0)

img_show("scharr-cabbage", img_scharr)
```
