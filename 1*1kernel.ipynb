```python
import torch
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
img = cv.imread("8.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()


img_rgb = np.array(img_rgb)
x = torch.from_numpy(img_rgb).float().unsqueeze(dim=0)

# tensor
print(x.shape)
```

```python
conv = torch.nn.Conv3d(in_channels=1,
                        out_channels=20,
                        kernel_size=(21,21,1),
                        stride=1)
x = conv(x)
print(x.shape)

# plt.imshow(x.detach().numpy().squeeze().astype('uint8'))
# plt.show()
```

```python
conv1 = torch.nn.Conv3d(in_channels=20,
                        out_channels=1,
                        kernel_size=(1,1,1),
                        stride=1)

x_1 = conv1(x)
print(x_1.shape)

plt.imshow(x_1.detach().numpy().squeeze().astype('uint8'))
plt.show()
```

```python
conv2 = torch.nn.Conv3d(in_channels=20,
                        out_channels=1,
                        kernel_size=(3,3,3),
                        stride=1)

x_3 = conv2(x)
print(x_3.shape)
plt.imshow(x_3.detach().numpy().squeeze().astype('uint8'))
plt.show()
```
