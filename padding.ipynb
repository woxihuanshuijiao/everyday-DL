```python
# 导入包
import torch
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# 我们现在可以叫它数据预处理

img = cv.imread("2.png", cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray")
# 转成numpy
img = np.array(img)


# 注意看！
# from_numpy这个操作经常用
# numpy转成tensor
x = torch.from_numpy(img).float().unsqueeze(dim=0)

# 看看张量形状
print(x.shape)
```

```python
# 定义一个卷积函数
show_padding = torch.nn.Conv2d(in_channels=1, out_channels=1, 
                            kernel_size=3, stride=1,
                            padding='same')

img_padding = show_padding(x)
print(img_padding.shape)

# 注意看，这个操作会经常用到！！！
# detach()返回一个新的tensor,是从当前计算图中分离下来的,但是仍指向原变量的存放位置
plt.imshow(img_padding.detach().numpy().squeeze(), cmap="gray")
```

```python
img2 = cv.imread("2.png", cv.IMREAD_GRAYSCALE)

#  注意看：这里[img] 已经升维
img2 = np.array([img2])

x2 = torch.from_numpy(img2).float()

print(x2.shape)

# 验证一下x和x2相等
print(x == x2)
```

```python
show_padding2 = torch.nn.Conv2d(in_channels=1, out_channels=1, 
                            kernel_size=3, stride=1,
                            padding='valid')

img_padding2 = show_padding2(x2)
print(img_padding2.shape)

plt.imshow(img_padding2.detach().numpy().squeeze(), cmap="gray")
```

```python
img3 = cv.imread("2.png", cv.IMREAD_GRAYSCALE)

#  注意看：这里[img] 已经升维
img3 = np.array([img3])

x3 = torch.from_numpy(img3).float()

print(x3.shape)
```

```python
# 偶数kernel
show_padding3 = torch.nn.Conv2d(in_channels=1, out_channels=1, 
                            kernel_size=4, stride=1,
                            padding='same')

img_padding3 = show_padding3(x3)
print(img_padding3.shape)

plt.imshow(img_padding3.detach().numpy().squeeze(), cmap="gray")
```
